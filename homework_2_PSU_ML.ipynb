```python
#Анализ датасета с киновзглядами 
df = pd.read_csv('Киновзгляды.csv')
import copy
originalDf = copy.deepcopy(df)   
```

```python
df
```

```python
titles = [ "Отметка времени","Мальчик или девочка?","Как часто тебя били в детстве?","Оцените ваш уровень вегетарианца по 5-ти бальной шкале","Что такое два.ч?","Твоя ориентация?","Погулять вечером после тяжкого рабочего дня, или посычевать дома?","Как часто играешь в игры?","Сериалы?","А книги? ","Вы таки антисемит? ","Какой предмет ближе всего по душе?","Как вы относитесь к аннексии Крыма?","Владимир Путин - ","Как хорошо вы знаете английский?","Было ли раньше лучше? ","А какие жанры фильмов любишь? "]
```

```python
import matplotlib.pyplot as plt
def showCounts(df,showColumn=[]) :
    for id,title in enumerate(titles[1:-1]):
        if len(showColumn)==0 or id+1 in showColumn  :
            df[title].value_counts().plot(kind='bar',figsize=(5,5))
            plt.suptitle(title+" "+str(id+1))
            plt.show()
 
 
```

```python
#Посмотрим статистику по каждому столбцу
showCounts(df)
```

```python
#Удалим лишнии столбцы и заменим строковые признаки 
df[titles[1]] = df[titles[1]] == 'Девочка'
df[titles[2]] = df[titles[2]] == 'Не били'
df[titles[4]] = df[titles[4]] == 'Не знаю'
df[titles[5]] = df[titles[5]] == 'Гетеро'
df[titles[6]] = df[titles[6]] == 'Гулять'
df[titles[10]] = df[titles[10]] == 'Нет'
df[titles[14]] = df[titles[14]] != 'Чуть лучше, чем ничего'
df.drop([titles[0],titles[13], titles[15], titles[16]], axis='columns', inplace=True)
df = pd.get_dummies(df, prefix=[titles[11], titles[12] ])
df
```

```python
#Посмотрим heatmap
import seaborn as sns
plt.figure(figsize=(14,12))
sns.heatmap(df.astype(float).corr(),linewidths=0.1, 
            square=True,  linecolor='white', annot=True)
plt.show()
```
